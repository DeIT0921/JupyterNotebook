# Hypothesis Testing 

#### in the real world, we want to make data-driven decisions, but we can't just do guess work to decide whether to do it or not.
#### that is where **Hypothesis Testing** comes in, to help us give a mathematically-backed way to make yes-no decisions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

## Introduction to t-Test 

```python
# Source: https://github.com/Opensourcefordatascience/Data-sets/blob/master/blood_pressure.csv
df = pd.read_csv("https://raw.githubusercontent.com/researchpy/Data-sets/master/blood_pressure.csv")
df.info()
```

```python
mbp = df[df['sex']=='Male']['bp_after']
fbp = df[df['sex']=='Female']['bp_after']
```

```python
print(round(mbp.mean(),2))
print(round(fbp.mean(),2))
print(round(mbp.mean()-fbp.mean(),2))
```
