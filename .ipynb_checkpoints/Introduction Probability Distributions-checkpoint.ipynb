# Probability Distributions

#### Discrite Uniform Distribution and Probability Mass Function

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import pandas as pd
```

```python
# membuat list berisi 6 angka yaitu [1-7]. Variabel ini merepresentasikan angka pada setiap sisi dadu
sisi_dadu = list(range(1,7))

# membuat probabilitas pada setiap sisi dadu
probabilitas_dadu = [1/6]*6

# membuat grafik dari probabilitas Mass Function
plt.figure(figsize=(12,8))
plt.bar(sisi_dadu, probabilitas_dadu)

# memberikan nama pada labels dari 1-6
plt.xticks(ticks=sisi_dadu, labels=['satu','dua','tiga','empat','lima','enam']);
```

```python
np.cumsum([1,2,3])
```

```python
plt.figure(figsize=(12,8))
# membuat grafik dengan cumulative distribution
plt.step(sisi_dadu, np.cumsum(probabilitas_dadu))
plt.xticks(ticks=sisi_dadu, label=['satu','dua','tiga','empat','lima','enam']);
```

## Bernoulli Distribution 

```python
# contoh: diberikan coin dimana ketika dilemparkan terdapat 70% kemungkinan mendapatkan sisi kepala atau gambar. 
sisi_coin = ['heads','tails']
probabilitas_coin = [0.7,0.3]

plt.figure(figsize=(10,8))
plt.bar(sisi_coin, probabilitas_coin)
```

## Binomial Distribution

```python
# note(.1 = 0.1 maka --> .x = 0.x)
probabilitas_heads = 0.7
n_flips = 50

# mendapatkan distribusi binomial
distribusi_coin = stats.binom(n_flips, probabilitas_heads)
kemungkinan_head = list(range(0,51))

# menampilkan dengan grafik
plt.figure(figsize=(10,8))
plt.bar(kemungkinan_head, distribusi_coin.pmf(kemungkinan_head),color='orange');
```

## Poisson Distribution 

```python
n_diners = 20

distribusi_diner = stats.poisson(n_diners)

kejadian_diners = list(range(50))
plt.figure(figsize=(10,8))
plt.bar(kejadian_diners, distribusi_diner.pmf(kejadian_diners), color='green')
```
