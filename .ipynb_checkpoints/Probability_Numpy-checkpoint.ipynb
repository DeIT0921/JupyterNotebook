# Introduction To Probability

```python
import numpy as np
```

```python
coin = ['heads','tails']
```

```python
np.random.choice(coin)
```

```python
count = 0
total_flips = 10000

for i in range(total_flips):
    if np.random.choice(coin) == 'heads':
        count+=1

print(count/total_flips)
```

```python
dice = [1,2,3,4,5,6]
```

```python
count = 0
total_throws = 10000

for i in range (total_throws):
    if np.random.choice(dice) == 2:
        count+=1

print(count/total_throws)
```
